```python
import pooch
import itk
from itkwidgets import view
```

```python
image_file = pooch.retrieve(url='https://w3s.link/ipfs/bafybeify6rxyhhl5ppgbwed6lq46zsqdz22lucthb4g5jyyahddk3paska/cthead1-bin.png',
                            known_hash='eaf0844db248ac949988932f626b834f8e995ab4534b3c746314b27e986c9e52')
```

```python
cthead1 = itk.imread(image_file)

view(cthead1)
```

Traditional distance transform. Faster but less accurate.

```python
distance = itk.signed_maurer_distance_map_image_filter(cthead1)

view(distance)
```

```python
parabolic_distance = itk.morphological_signed_distance_transform_image_filter(cthead1)

view(parabolic_distance)
```
